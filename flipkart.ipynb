```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as rq
```

```python
url = "https://www.flipkart.com/search?q=mobile%20under%2010000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
data = rq.get(url)
response = data.text
```

```python
soup = BeautifulSoup(response,"html.parser")
```

```python
print(soup.prettify())
```

```python
box = soup.find("div","DOjaWF gdgoEp")
```

```python
next_page = box.find("a",class_="_9QVEpD").get("href")
```

```python
print(next_page)
```

```python
list_url=[]
```

```python
for index in range(2,25):
    page_url = f"https://www.flipkart.com/search?q=mobile+under+10000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={index}"
    list_url.append(page_url)
    print(page_url)
print(len(list_url))
```

```python
product_name_list= []
price_list = []
description_list = []
review_list = []
```

```python
product_name =  box.find_all('div',class_ = 'KzDlHZ')
```

```python
for product_data in product_name:
    p_data = product_data.text
    product_name_list.append(p_data)
```

```python
product_name_list
```

```python
prices = box.find_all("div",class_="Nx9bqj _4b5DiR")
```

```python
for price in prices:
    price_data = price.text
    price_list.append(price_data)
```

```python
price_list
```

```python
descriptions = box.find_all("ul",class_ = "G4BRas")
```

```python
for decription in descriptions:
    des = decription.text
    description_list.append(des)
```

```python
description_list
```

```python
print(len(description_list))
```

```python
reviews = box.find_all("div",class_ = "XQDdHH")
```

```python
for review in reviews:
    review_data = review.text
    review_list.append(review_data)
```

```python
review_list
```

```python
print(len(review_list))
```

```python
df = pd.DataFrame({"Product": product_name_list,"Description":description_list,"Review":review_list})   
```
